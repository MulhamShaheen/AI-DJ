```python
path_casual = 'casual'
path_sport = 'sport'
path_formal = 'formal'
```

```python
#переименование файлов
def rename(directory, name):
    for count, f in enumerate(os.listdir(directory)):
        f_name, f_ext = os.path.splitext(f)
        f_name = name + '_'+str(count)
 
        new_name = f'{f_name}{f_ext}'
        os.rename(os.path.join(directory, f), os.path.join(directory, new_name))
```

```python
#перемешаем списки файлов
list_files_shuffled_casual = os.listdir('casual')
list_files_shuffled_sport = os.listdir('sport')
list_files_shuffled_formal = os.listdir('formal')
```

```python
random.shuffle(list_files_shuffled_casual)
random.shuffle(list_files_shuffled_sport)
random.shuffle(list_files_shuffled_formal)
```

```python
#создаем отдельные файлы для создания датасетов train, eval, test для сохранения сбалансированности данных
list_files_train = list_files_shuffled_casual[0:700] + list_files_shuffled_sport[0:700] + list_files_shuffled_formal[0:700]
list_classes_train = [0] * 700 + [1] * 700 + [2] * 700 
df_files = pd.DataFrame(
    {'files': list_files_train,
     'classes': list_classes_train
    })
df_files.to_csv('files_train.csv')
```
